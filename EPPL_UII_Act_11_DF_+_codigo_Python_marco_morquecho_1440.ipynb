```python
# Programa para calcular el area y perimetro de un triangulo
# Descripcion: Calcula el area

import math

print("calcular area y perimetro de un triangulo")

# Entrada
b = float(input("ingrese la base: "))
h = float(input("ingrese la altura: "))

a = float(input("ingrese el lado a: "))
b2 = float(input("ingrese el lado b: "))
c = float(input("ingrese el lado c: "))

# Proceso
A = (b * h) / 2
P = a + b2 + c

# Salida
print("el area del triangulo es: ", A)
print("el perimetro del triangulo es: ", P)
print("marco morquecho NL 39")
```
